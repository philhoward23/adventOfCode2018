```python
#strategy: keep track of points where stream falls directly down
#once hit clay, explore left and right until one or two new points are found
#repeat with new points
```

```python
sources = [(500,0)]
```

```python
def fill_down(start,end,space):
    space[start[0],start[1]:end[1]] = 2
    return(0)

def fill_left(start,end,space):
    space[end[0]+1:start[0],end[1]] = 2
    return(0)

def fill_right(start,end,space):
    space[start[0]:end[0],end[1]] = 2
    return(0)

```

```python
def find_next_clay(source,scanspace):
    next_clay = np.where(scanspace[source[0],source[1]:] >= 1)[0]
    if len(next_clay) > 0:
        #pretend clay below water if water was found first
        if scanspace[source[0],source[1]+min(next_clay)] == 2:
            next_clay = min(next_clay) + 1
        else:
            next_clay = min(next_clay)
        next_clay = (source[0],source[1]+next_clay)
    else:
        #no other clay below so set to max of space
        next_clay = (source[0],len(scanspace[source[0]]))
    fill_down(source,next_clay,scanspace)
    return(next_clay)
```

```python
def blocked_left_right(blocked,scanspace):
    blocked_left = np.where(scanspace[:blocked[0],blocked[1]-1] == 1)[0]
    if len(blocked_left) > 0:
        blocked_left = (max(blocked_left),blocked[1]-1)
    else:
        #no other clay to the left so set to min of space
        blocked_left = (0,blocked[1]-1)
    blocked_right = np.where(scanspace[blocked[0]+1:,blocked[1]-1] == 1)[0]
    if len(blocked_right) > 0:
        blocked_right = (blocked[0]+1+min(blocked_right),blocked[1]-1)
    else:
        #no other clay to the right so set to max of space
        blocked_right = (len(scanspace),blocked[1]-1)
    return(blocked_left, blocked_right)

def fall_left_right(blocked,scanspace):
    #either the next zero value on row below, or the next water on this level (falling)
    next_left = max(np.where(scanspace[:blocked[0],blocked[1]] == 0)[0])
    if scanspace[next_left+1,blocked[1]] == 2:
        next_left = next_left+1  
    fall_left = (next_left,blocked[1]-1)
    #either the next zero value on row below, or the next water on this level (falling)
    next_right = blocked[0]+1+min(np.where(scanspace[blocked[0]+1:,blocked[1]] == 0)[0])
    if scanspace[next_right-1,blocked[1]] == 2:
        next_right = next_right-1
    fall_right = (next_right,blocked[1]-1)   
    return(fall_left, fall_right)
```

```python
def flow_left_right(blocked,scanspace):
    blocked_left, blocked_right = blocked_left_right(blocked,scanspace)
    fall_left, fall_right = fall_left_right(blocked,scanspace)
    fill_left(blocked,(max(fall_left[0],blocked_left[0]),blocked[1]-1),scanspace)
    fill_right(blocked,(min(fall_right[0],blocked_right[0]),blocked[1]-1),scanspace)
    #print(blocked_left, blocked_right, fall_left, fall_right)
    return(blocked_left, blocked_right, fall_left, fall_right)
```

```python
sources = [(500,0)]

while len(sources) > 0:
    found = False
    source = sources.pop()
    print('next source at ', source, ':', len(sources), ' sources left')
    blocked = find_next_clay(source,scanspace)
    #skip if no next clay found
    if blocked[1] < len(scanspace[blocked[0]]):
        while not found:
            #flow water and raise level if no new sources found
            blocked_left, blocked_right, fall_left, fall_right = flow_left_right(blocked,scanspace)
            if fall_left[0] > blocked_left[0]:
                found = True
                if scanspace[fall_left[0],fall_left[1]] == 0:
                    sources.append(fall_left)
            if fall_right[0] < blocked_right[0]:
                found = True
                if scanspace[fall_right[0],fall_right[1]] == 0:
                    sources.append(fall_right)
            if not found:
                #raise water
                blocked = (blocked[0],blocked[1] - 1)

        
```
